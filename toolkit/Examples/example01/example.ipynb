We start by importing the postprocessing `vaspout_h5` module into our script, and creating objects for the Band Structure (BS) and Density of States (DOS) calculations.

```python
from vaspout_h5 import vaspout_h5, plot_BS_DOS
import matplotlib.pyplot as plt

BS = vaspout_h5('/home/karol/Monolayers/T_Wozniak/h5_post/np_bs/vaspout.h5')
DOS = vaspout_h5('/home/karol/Monolayers/T_Wozniak/h5_post/np_dos/vaspout.h5')
```

The easiest thing we can do is to simply call an appropriate plotting method for a class, so let's try it!

```python
BS.plot_BS()
plt.show()
```

The result seems to be pretty unclear. The easiest way to get around that is to call the default plotting method

```python
BS.plot_BS_default()
plt.show()
```

This will have also saved that image to our BS vaspout's folder.
These are regular matplotlib plots, so we can modify them however we want. Let's try to make the first plot more informative:

```python
plt.figure(figsize=(4,6))
BS.plot_BS()
plt.ylim([-7.5,10])
plt.show()
```

We can further modify it in various ways:

```python
fig, axes = plt.subplots(ncols=3, sharey=True, figsize=(12,6))
plt.ylim([-7.5,10])
BS.plot_BS('clean', bands='25-28, 39, 40', ax=axes[0], bandnums=True, E0='fermi')
BS.plot_BS('Mo d',ax=axes[1], bandnums=True, alpha = .05)
BS.plot_BS('Te p',ax=axes[1], alpha = .05)
BS.plot_BS('2 Mo x dxy', kpaths='1 2 -2 -1', ax=axes[2], E0=0)
plt.show()
```

Similarly with DOS plots:

```python
fig, axes = plt.subplots(ncols=3, sharey=True, figsize=(12,6))
DOS.plot_DOS('total',ax=axes[0])
DOS.plot_DOS('Mo d',ax=axes[1])
DOS.plot_DOS('Te p',ax=axes[1])
DOS.plot_DOS('2 Mo dxy',ax=axes[2])
plt.show()
```

And to easily make a cohesive BS+DOS plot we have the `plot_BS_DOS` function:

```python
plot_BS_DOS(BS, DOS, 'Mo d')
```
